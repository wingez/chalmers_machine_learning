```python
import numpy as np
import matplotlib.pyplot as plt

from system.model import simulate
```

```python

result = simulate(seconds=0.2, target_current=3)



#plt.subplot(5,1,1)
plt.plot(result.time,result.current,)
plt.xlabel("tid")
plt.ylabel("str√∂m")
plt.show()
```

```python
plt.plot(result.time,result.omega,)
plt.xlabel('tid')
plt.ylabel("omega")
plt.show()
```

```python
plt.plot(result.time,result.pos,)
plt.xlabel('tid')
plt.ylabel("pos")
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
#fig,ax = plt.subplots(figsize = (100,100))
for label,voltage in result.voltages.items():

  ax.scatter(result.time, voltage, label= label)

ax.set_xlabel("time (s)")
ax.set_ylabel("voltage (V)")
ax.legend()
```

```python
fig = plt.figure()
ax = plt.axes()
#fig,ax = plt.subplots(figsize = (100,100))
ax.plot(result.time, result.t_dev, label = "t_dev")
ax.plot(result.time, result.t_last, label = "t_last")

ax.set_xlabel("time (s)")
ax.set_ylabel("torque (N)")
ax.legend()




#fig.show()
#plt.show()

```
